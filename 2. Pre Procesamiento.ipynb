```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.min_rows', 6)
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('jugadores.csv')
```

```python
df
```

```python
#Obtener numero de jornada
JCol=[]
for j in range (1, 39):
    jornadas = [j for i in range(1,221)]
    JCol.append(jornadas)
import itertools
JCol = list(itertools.chain.from_iterable(JCol))
df['jornada'] = JCol
```

```python
#Replace NaN to 0
df.replace('-',0, inplace=True)
df.replace(np.nan,0, inplace=True)
```

```python
#Clean Titulares strings
df['Titulares'] = df['Titulares'].str.replace('\d+', '')
df['Titulares'] = df['Titulares'].str.replace(" '", '')
```

```python
df
```

```python
#Cambiar jugadores a numeros
keys = df['Titulares'].unique()
values = range(len(keys))
dicc_jug = dict(zip(keys,values))
df['Titulares'] = df['Titulares'].replace(dicc_jug)
```

```python
#Seleccionar los atributos que vamos a usar
df = df[['Titulares', 'T', 'TaP', 'C', 'CP', 'CF', 'Reg', 'PA', 'partido',
       'equipo', 'jornada']]
```

```python
#Pasar los corners al final
cf = df['CF']
df.drop(labels=['CF'], axis=1,inplace = True)
df.insert(10,'CF', cf)
```
